```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# 1. Load data with the correct encoding
df = pd.read_csv('spam.csv', encoding='ISO-8859-1')

# 2. Preprocessing: Drop junk columns and rename
df = df.drop(columns=df.columns[2:5])
df.columns = ['Category', 'Message']
df['Spam'] = df['Category'].apply(lambda x: 1 if x == 'spam' else 0)

# 3. Visualizing Data Distribution (FIXED to remove FutureWarning)
plt.figure(figsize=(12, 5))

# Subplot 1: Bar Chart
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Category', hue='Category', palette='viridis', legend=False)
plt.title('Count of Spam vs Ham Messages')

# Subplot 2: Pie Chart
plt.subplot(1, 2, 2)
df['Category'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#66b3ff','#ff9999'], startangle=90, explode=(0, 0.1), shadow=True)
plt.title('Percentage of Spam')
plt.show()

# 4. Feature Engineering: Message Length Analysis
df['Length'] = df['Message'].apply(len)
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='Length', hue='Category', bins=50, kde=True, palette='magma')
plt.title('Message Length Distribution: Ham vs Spam')
plt.show()

# 5. Model Building Pipeline
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.Spam, test_size=0.25, random_state=42)

# Professional Pipeline: Vectorizer -> TF-IDF -> Naive Bayes
pipeline = Pipeline([
    ('vectorizer', CountVectorizer(stop_words='english')),
    ('tfidf', TfidfTransformer()),
    ('classifier', MultinomialNB())
])

pipeline.fit(X_train, y_train)

# 6. Evaluation and Heatmap
predictions = pipeline.predict(X_test)
print(f"Final Model Accuracy: {accuracy_score(y_test, predictions) * 100:.2f}%")
print("\nClassification Report:\n", classification_report(y_test, predictions))

# Confusion Matrix Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, predictions), annot=True, fmt='d', cmap='Blues', xticklabels=['Ham', 'Spam'], yticklabels=['Ham', 'Spam'])
plt.title('Confusion Matrix: Spam Detection')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# 7. Testing with custom inputs
def predict_message(text_list):
    results = pipeline.predict(text_list)
    for msg, res in zip(text_list, results):
        status = "SPAM" if res == 1 else "HAM (Safe)"
        print(f"Message: {msg}\nResult: {status}\n{'-'*20}")

test_msgs = ["Get a free $1000 prize now!", "Hey, are you coming to the office today?"]
predict_message(test_msgs)
```
